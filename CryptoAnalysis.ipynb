```python
#Importing dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Preparation

```python
file = 'crypto_data.csv'
df = pd.read_csv(file, index_col=0)
df
```

```python
#Discard all cryptocurrencies that are not being traded. 
#Once you have done this, drop the IsTrading column from the dataframe.

clean_df = df[df['IsTrading']!=False]
clean_df.drop(columns = ['IsTrading'], inplace=True)
clean_df
```

```python
#Remove all rows that have at least one null value.
clean_df.dropna(inplace=True)
clean_df
```

```python
#Filter for cryptocurrencies that have been mined. That is, the total coins mined should be greater than zero.
clean_df = clean_df[clean_df['TotalCoinsMined']>0]
clean_df
```

```python
#delete the CoinName from the original dataframe.
clean_df.drop(columns = ['CoinName'], inplace=True)
clean_df
```

```python
#use Pandas to create dummy variables. 
dummy_df = pd.get_dummies(clean_df, columns=['Algorithm', 'ProofType'])
dummy_df
```

#### Examine the number of rows and columns of your dataset now. How did they change?

The number of rows did not change with get dummies, as expected. However, the number of columns increased significantly, from 4 to 98. That is a massive amount of features for the model to digest, so we will move onto scaling the data and performing a dimensionality reduction with PCA.

```python
#Standardize your dataset so that columns that contain larger values do not unduly influence the outcome.
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.decomposition import PCA

scaled_data = StandardScaler().fit_transform(dummy_df)

scaled_data
```

```python
#Perform dimensionality reduction with PCA.
pca = PCA(n_components=0.90)

pca_out = pca.fit_transform(scaled_data)

pca_out
```

```python
pca_out.shape
```

####  How did the number of the features change?
The PCA reduction brought the number of features down from 98 to 74.

```python
#further reduce the dataset dimensions with t-SNE and visually inspect the results.
from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=100)    

tsne_out = tsne.fit_transform(pca_out)

tsne_out
```

```python
tsne_out.shape
```

```python
#create a scatter plot of the t-SNE output. Observe whether there are distinct clusters or not.
X = tsne_out[:,0]
y = tsne_out[:,1]

plt.scatter(X, y)
plt.show()
```

There are most certainly distinct clusters that have emerged from the PCA and t-SNE reductions.

```python
#Create an elbow plot to identify the best number of clusters. 
#Use a for-loop to determine the inertia for each k between 1 through 10. 
#Determine, if possible, where the elbow of the plot is, and at which value of k it appears.
from sklearn.cluster import KMeans

inertia = []
k = list(range(1, 10))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(tsne_out)
    inertia.append(km.inertia_)

elbow = {"k": k, "inertia": inertia}
elbow_df = pd.DataFrame(elbow)

plt.plot(elbow_df['k'], elbow_df['inertia'])
plt.xticks(list(range(10)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
